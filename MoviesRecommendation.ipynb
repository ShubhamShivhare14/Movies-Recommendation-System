```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)
pd.set_option('display.expand_frame_repr', False)

movies = pd.read_csv(r'C:\Users\shubham\Desktop\movies recommendation\movies.csv')
ratings = pd.read_csv(r'C:\Users\shubham\Desktop\movies recommendation\ratings.csv')
df = movies.merge(ratings, how="left", on="movieId")
```

```python
# Replace missing values
df.dropna(inplace=True)

# if a movie is rated less, lets remove those movies from the dataset
rating_counts = pd.DataFrame(df['title'].value_counts())
rare_movies = rating_counts[rating_counts['title']<=50].index
common_movies = df[~df['title'].isin(rare_movies)]

common_movies.head()
```

```python
common_movies.shape
```

```python
user_movie_df = common_movies.pivot_table("rating", "userId", "title")

user_movie_df.head()
```

```python
# I am selecting a random user for whom I am going to generate recommendation
random_user = int(pd.Series(user_movie_df.index).sample(1, random_state=20).values)

random_user_df = user_movie_df[user_movie_df.index == random_user]

random_user_df.head()
```

```python
movies_watched = random_user_df.columns[random_user_df.notna().any()].tolist()
```

```python
movies_watched
```

```python
len(movies_watched)
```

```python
# We are selecting the status of all users with 47 above movies
movies_watched_df = user_movie_df[movies_watched]
```

```python
movies_watched_df.head()
```

```python
# Interested user -- 121
# Movies watched -- 47

# Find all users who have watched some of these 47 movies
# 1 -- 14
# 2 --  1

# Remove users who have watched less common movies --> Agree with this or not? 
```

```python
# find out the count of common movies that every user have watched.
# Remove where the count is less then 20

user_movie_count = movies_watched_df.T.notnull().sum()
user_movie_count = user_movie_count.reset_index()
user_movie_count.columns = ['userId', 'movie_count']
user_movie_count.head()
```

```python
# Remove where the count is less then 20
users_same_movies = user_movie_count[user_movie_count['movie_count'] > 20]['userId']

users_same_movies.head()
```

```python
final_df = pd.concat([movies_watched_df[movies_watched_df.index.isin(users_same_movies)], 
                     random_user_df[movies_watched]])

final_df.head()
```

```python
final_df.shape
```

```python
# We are taking the correlations between the movies these users watched and the ratings they gave for each movie.
corr_df = final_df.T.corr().unstack().sort_values().drop_duplicates()

corr_df = pd.DataFrame(corr_df, columns=["corr"])

# We are renaming the index names of corr_df as user_id_1 and user_id_2.
corr_df.index.names = ['user_id_1', 'user_id_2']

corr_df = corr_df.reset_index()

corr_df.head()
```

```python
# We select the users whose correlation with the random user is at least 0.50.
top_users = corr_df[(corr_df["user_id_1"] == random_user) & (corr_df["corr"] >= 0.50)][
    ["user_id_2", "corr"]].reset_index(drop=True)

top_users = top_users.sort_values(by='corr', ascending=False)

top_users.rename(columns={"user_id_2": "userId"}, inplace=True)

top_users.head()
```
